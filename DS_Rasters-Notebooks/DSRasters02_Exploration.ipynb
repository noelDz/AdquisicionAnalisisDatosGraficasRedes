# Exploración del DataFrame

En este notebook, utilizaremos algunas herramientas de Pandas para explorar el registro simultáneo de las células. Para esto, primero cargaremos el archivo preprocesado en el Notebook anterior:

```python
rawdata = pd.read_csv('./CellsData_pd.csv',index_col=0)
nRows = rawdata.shape[0]
nCells = rawdata.shape[1]
rawdata.head()
```

## 2 ¿ Qué valores que toma el DataFrame?

Para realizar un análisis adecuado, es muy importante tomar en cuenta el rango de valores que toma cada columna de nuestra tabla, esto nos ayudará más adelante a elegir los métodos más adecuados para realizar análisis cuantitativos.

Ahora exploraremos los valores que toma cada columna. Para ver los valores diferentes que toma la primer columna, utilizaremos lo siguiente:

```python
rawdata['cell_1'].unique()
```

Podemos ver que los valores que toma son **0**, correspondiente a la célula **en reposo** y **1, correspindiente a la célula presentando un potencial de acción**.

Podemos hacer esta misma exploración para el resto de las células:

```python
unique_values = [rawdata[cell].unique() for cell in rawdata.columns]
unique_values
```

## 3 Descriptivos Estadísticos 
Para obtenet los descriptivos generales de un DataFrame, basta con usar el método **.describe()**:

```python
description_df =rawdata.describe()
description_df
```

## 4 Conteo de Spikes
Para acceder al número de unos y ceros que contiene cada columna es posible usar el método **.value_counts( )** :

```python
rawdata['cell_15'].value_counts()
```

Podemos ver que la **célula 15** presenta 135 spikes durante el registro.
Obtendremos estos mismos conteos para el resto de las células:

```python
val_counts = [rawdata[cell].value_counts() for cell in rawdata.columns]
val_counts[0] # Conteo de valores para la célula 1.
```

### 4.1 Tabla resumen de la actividad celular

Ahora crearemos una tabla que incluya el número de 0's y 1's por célula, además del resumen estadístico:

```python
cell_summary = [val_counts[i].append(description_df.iloc[:,i]) for i in range(nCells)]
cell_summary
```

### 4.2 Convertir una lista o arreglo a DataFrame
Ahora, convertiremos a DataFrame la lista anterior:

```python
cell_summary_df = pd.DataFrame(cell_summary)
cell_summary_df.head()
```

# 5 Filtrando columnas 
Finalmente, algo que podemos hacer para tener un conjunto de datos más limpio, es eliminar aquellas células que no hayan presentado *spikes* durante el registro.

Para esto, primero contaremos el total de *spikes* que realizó cada célula:

```python
rawdata['cell_1'].sum()
```

```python
#Primero sumamos el total de spikes por cada célula (columna)
spikes_cell = [rawdata[cell].sum() for cell in rawdata.columns]

#... y convertimos la lista anterior a un arreglo:
spikes_cell = np.asarray(spikes_cell)
spikes_cell
```

### 5.1 Células Activas e Inactivas

```python
inactiveCells = sc.where(spikes_cell==0)[0]
activeCells = sc.where(spikes_cell>0)[0]

print("El número de células activas es: %s" % (len(activeCells)))
print("El número de células inactivas es: %s" % (len(inactiveCells)))
```

La lista de células inactivas es la siguiente:

```python
inactiveCells
```

### 5.2 Método .drop( )
Ahora, quitaremos las células inactivas y guardaremos el resultado en la nueva variable *workdata*, para esto utilizaremos el método **.drop( )**.

```python
workdata_active = rawdata.drop(rawdata.columns[inactiveCells], axis=1)
workdata_active.shape
```

```python
workdata_active.to_csv('CellsData_clean.csv')
```
